***EDA***

For each variable in dataset we will find:-
1- Is the variable contiuous or categorical?
2- Are there any missing values?
3-Min,Max,Mean and Standard Deviation for continuous variable
4-Histogram describing distribution of each variable


***EDA for PassengerID***

```python
df.PassengerId.describe()
```

We see that for passengerID we are able to calculate values like mean and standard deviation.
Hence ***PassengerID is a continuous variable***

```python
df[df.PassengerID.isnull()]
```

```python
df[df.PassengerId.isnull()]
```

There are ***no missing values*** in PassengerID

```python
df.PassengerId.describe()
```

The above output gives us the ***min,max,mean and standard Deviation*** values for PassengerID

```python
passenger_Histogram=df.PassengerId.hist(bins=20)
passenger_Histogram.set_title("Passenger Histogram")
```

The graph above is ***Histogram For Passenger***

***EDA for Survived***


```python
df.Survived.value_counts()
```

The above output shows that the output is divided into 2 categories Male & Female. Thus ***Survived is a Categorical Variable***

```python
df[df.Survived.isnull()]
```

The above output shows that there are ***No Missing Values in the Survived variable***

```python
survived_Histogram=df.Survived.hist(bins=10)
survived_Histogram.set_title("Survived Histogram")
survived_Histogram.set_xlabel("Category: 0- Dead 1-Survived")
survived_Histogram.set_ylabel("Passenger Count")
```

We can thus represent the data of Survived Variable in ***Survived Histogram***
