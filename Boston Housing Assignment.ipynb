# Boston Housing Assignment

Implimenting scikit learn's r2 score and mse methods for measuring the performance of the linear regressor and sklearn.linear_model.Ridge modle

```python
from sklearn import datasets
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cross_validation import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

```python
bean = datasets.load_boston()
print bean.DESCR
```

```python
def load_boston():
    scaler = StandardScaler()
    boston = datasets.load_boston()
    X=boston.data
    y=boston.target
    X = scaler.fit_transform(X)
    return train_test_split(X,y)
```

```python
X_train, X_test, y_train, y_test = load_boston()
```

```python
X_train.shape
```

## Fitting a Linear Regression

This is done by instantiating a new regression object (line 1) and giving your regression object your training data (line 2) by calling .fit(independent variables, dependent variable)

```python
clf = LinearRegression()
clf.fit(X_train, y_train)
```

## Making a Prediction

```python
zip (y_test, clf.predict(X_test))
```

```python
mean_squared_error(y_test,clf.predict(X_test))
```

```python
r2_score(y_test,clf.predict(X_test))
```

## MSE = 24.335492657229299

## r2_score = 0.74715765225195818

## Computing Linear Regression

taking alpha=0.001 for computing sklearn.linear_model.Ridge


```python
from sklearn.linear_model import Ridge
r = Ridge(alpha=.001)
r.fit(X_train, y_train)
```

```python
zip (y_test, r.predict(X_test))
```

```python
mean_squared_error(y_test,r.predict(X_test))
```

```python
r2_score(y_test,r.predict(X_test))
```

## MSR = 24.335503991903398

## r2_score = 0.74715753448628763


# Ridge Linear Regression



### taking alpha =2

```python
from sklearn.linear_model import Ridge
r = Ridge(alpha=2)
r.fit(X_train, y_train)
```

```python
zip (y_test, r.predict(X_test))
```

```python
mean_squared_error(y_test,r.predict(X_test))
```

```python
r2_score(y_test,r.predict(X_test))
```

## MSR = 24.361937550142731

## r2_score = 0.74688289353207704

Thus we see that when value of alpha is decreased to 0.001 there is not much difference in MSR and R2 values.
When we change the value of alpha to 2 the value of MSR changes very slightly.However R2 chanes a bit
